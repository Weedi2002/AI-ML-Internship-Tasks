```python
# Step 1: Import Required Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the Iris Dataset
# Load from seaborn
iris = sns.load_dataset('iris')
```

```python
# Step 3: Basic Data Inspection
print("✅ Dataset Loaded Successfully!\n")
print("Shape of dataset:", iris.shape)
print("\nColumn Names:", iris.columns.tolist())
print("\nFirst 5 Rows:")
display(iris.head())
```

```python
# Step 4: Summary Statistics
print("\nDataset Information:")
print(iris.info())

print("\nSummary Statistics:")
display(iris.describe())
```

```python
# Step 5: Check for Missing Values
print("\nMissing Values per Column:")
print(iris.isnull().sum())
```

```python
# Step 6: Data Visualization
# 1. Scatter Plot — relationships between features
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', palette='Set2')
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.show()
```

```python
# 2. Histogram — value distribution
iris.hist(figsize=(10,8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Iris Dataset Features', fontsize=14)
plt.show()
```

```python
# 3. Box Plot — detecting outliers
plt.figure(figsize=(10,6))
sns.boxplot(data=iris)
plt.title('Box Plot of All Features')
plt.show()
```

```python
# 4. Pair Plot — visualize relationships across all features
sns.pairplot(iris, hue='species', palette='husl')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
# Step 7: Insights
print("\n📊 Final Insights:")
print("""
1. The Iris dataset contains 150 samples with 4 numerical features and 1 categorical target (species).
2. There are no missing values — the data is clean.
3. Petal length and petal width show strong separation between species.
4. Sepal features overlap more between species, making them less discriminative.
5. Distributions are generally symmetric with few potential outliers.
""")
```
